```python
data_student={'Name':['radha','trisha','hina','gina','elena'],'Science':[100,40,np.nan,78,88],'Maths':[80,50,170,56,79],'English':[85,78,33,np.nan,120]}
```

```python
df['Science']=df['Science'].fillna(df['Science'].mean())
```

```python
df['English']=df['English'].fillna(df['English'].median())
```

```python
df
```

```python
for column in ['Science','Maths','English']:
  sns.boxplot(df[column])
  plt.show()
```

```python
def handel_outlier(column):
  Q1=df[column].quantile(0.25)
  Q3=df[column].quantile(0.75)
  IQR= Q3-Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  df[column] = np.where(df[column] < lower_bound, lower_bound,
  df[column])
  df[column] = np.where(df[column] > upper_bound, upper_bound,
  df[column])
```

```python
for column in ['Science','Maths','English']:
  handel_outlier(column)
```
